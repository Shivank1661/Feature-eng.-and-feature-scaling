# Filling Missing values

```python
#From the above observation column Product_Category_2 it clearly shows that it is normally distributed.
#the null values in column Product_Category_2 is filled with any measure central tendency(mean,median,mode) beacause it is normally distributed
```

```python
df["Product_Category_2"].mode()[0]
```

```python
df["Product_Category_2"].unique()
```

```python
#replace the missing value with mode
df["Product_Category_2"]=df["Product_Category_2"].fillna(df["Product_Category_2"].mode()[0])
```

```python
df["Product_Category_2"]
```

```python
#From the above observation column Product_Category_3 it clearly shows that it is -ve skewed distributed.
#the null values in column Product_Category_3 is filled with mode beacause it is -ve skewed distributed
```

```python
df["Product_Category_3"].mode()
```

```python
#replace the missing value with mode
df["Product_Category_3"]=df["Product_Category_3"].fillna(df["Product_Category_3"].mode()[0])
```
